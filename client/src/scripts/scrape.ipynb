```python
from bs4 import BeautifulSoup
import requests


def get_url(url):
    r = requests.get(url).text
    soup = BeautifulSoup(r)
    return soup


html = get_url('https://www.imdb.com/list/ls052283250/')
```

```python
imgs = html.find_all('div', {'class': 'lister-item-image'})

#print(imgs)
import json
data = []
for img in imgs:
    data.append({'image': img.find('img')['src'], 'name': img.find('img')['alt']})

with open('data.json', 'w') as f:
    json.dump(data, f, ensure_ascii=False)


```
